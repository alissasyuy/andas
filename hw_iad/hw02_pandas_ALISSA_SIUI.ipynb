11.1) Для каждого уровня образования выведите минимальный балл за экзамен по письму


```python
%%time
mydict = {"associate's degree": list(), "bachelor's degree": list(), 'high school': list(),
              "master's degree": list(),'some college': list(), 'some high school': list()}
for i in df.index:
  mydict[df.loc[i]['education']].append(df.loc[i]['writing score'])

for i in mydict:
  value = mydict[i]
  mydict[i] = min(value)
mydict
```

11.2) Для каждой этнической группы выведите средний балл за экзамен по чтению немного другим способом

```python
%%time
# заносим значения в словарь
mydict = dict()
for i in df.index:
  if df.loc[i]['race/ethnicity'] in mydict:
    mydict[df.loc[i]['race/ethnicity']].append(df.loc[i]['reading score'])
  else:
    mydict[df.loc[i]['race/ethnicity']] = [df.loc[i]['reading score']]

# считаем среднее
for i in mydict:
  value = mydict[i]
  mydict[i] = sum(value)/len(value)
dict(sorted(mydict.items()))
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
new_df = df[['math score', 'reading score', 'writing score']] # выбираем базу расчета
df['OverAll_PassStatus'] = (new_df>=50).prod(axis = 1) # перемножаем построчно булевые
df['OverAll_PassStatus'].replace({1: 'P', 0: 'F'}, inplace=True)
df
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(x):
  hhh = ''
  if 0 <= x < 50:
    hhh = 'FAIL'
  if 50<= x <60:
    hhh = 'E'
  if 60<= x <70:
    hhh = 'D'
  if 70<= x <80:
    hhh = 'C'
  if 80<= x <90:
    hhh = 'B'
  if 90<=x:
    hhh = 'A'
  return hhh
df['Grades'] = df[df.columns[-7:-4]].mean(axis = 1, numeric_only=True).apply(lambda x: GetGrade(x))
df
```

```python
pd.DataFrame(df['Grades'].value_counts()).sort_values('Grades', ascending=False)
```

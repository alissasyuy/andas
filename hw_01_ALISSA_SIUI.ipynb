- –ù–∞–π–¥–∏—Ç–µ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –ø–æ–¥–º–∞—Ç—Ä–∏—Ü—ã –º–∞—Ç—Ä–∏—Ü—ã $X$, —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–Ω–æ–π –Ω–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–∏ 2 –∏ 3 —Å—Ç—Ä–æ–∫–∏ –∏ 1 –∏ 2 —Å—Ç–æ–ª–±—Ü–∞; —Å—á–∏—Ç–∞–µ–º, —á—Ç–æ —Å—Ç—Ä–æ–∫–∏ –∏ —Å—Ç–æ–ª–±—Ü—ã –Ω—É–º–µ—Ä—É—é—Ç—Å—è —Å –µ–¥–∏–Ω–∏—Ü—ã (–∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ slicing!). –¢–∞–∫–æ–π –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è **–º–∏–Ω–æ—Ä–æ–º** –º–∞—Ç—Ä–∏—Ü—ã $X$.
- –ù–∞–π–¥–∏—Ç–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ $X^TX$, –µ–≥–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –∏ —Å–ª–µ–¥.

```python
matrix_prod = (X.T).dot(X)
det_matrix_prod = np.linalg.det(matrix_prod)
sled_matrix_prod = np.sum(np.diag(matrix_prod))

print(f"X^T*X –Ω–æ–≤–∞—è –º–∞—Ç—Ä–∏—Ü–∞ = {matrix_prod}")
print(f"–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –Ω–æ–≤–æ–π –ø–æ–¥–º–∞—Ç—Ä–∏—Ü—ã = {det_matrix_prod.round(2)}")
print(f"–°–ª–µ–¥ –Ω–æ–≤–æ–π –º–∞—Ç—Ä–∏—Ü—ã = {sled_matrix_prod}")
```

–ù–∞ –ø–∞—Ä–µ –º—ã –≥–æ–≤–æ—Ä–∏–ª–∏, —á—Ç–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ü–∏–∫–ª–æ–≤ (–∏ —Å–∞–º–æ–ø–∏—Å–Ω—ã—Ö –≤—ã—á–∏—Å–ª–µ–Ω–∏–π –≤ —Ü–µ–ª–æ–º) –≤ python –ª—É—á—à–µ –ø–æ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –∏–∑–±–µ–≥–∞—Ç—å –∏ –≤–∞–∂–Ω–æ —É–º–µ—Ç—å –Ω–∞—Ö–æ–¥–∏—Ç—å —Å–ø–æ—Å–æ–±—ã –¥–µ–ª–∞—Ç—å –≤—Å—ë –±–∏–±–ª–∏–æ—Ç–µ—á–Ω—ã–º–∏ —Å—Ä–µ–¥—Å—Ç–≤–∞–º–∏.

–í —Å–ª–µ–¥—É—é—â–∏—Ö —Ç—Ä—ë—Ö –∑–∞–¥–∞–Ω–∏—è—Ö –Ω–µ–ª—å–∑—è –ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —Ü–∏–∫–ª–∞–º–∏ –≤–æ–æ–±—â–µ, –∞ —Ç–∞–∫–∂–µ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏—è–º–∏ `map` –∏ `reduce` –∏ –∏–º –ø–æ–¥–æ–±–Ω—ã–º–∏; –≤–º–µ—Å—Ç–æ —ç—Ç–æ–≥–æ –ø–æ—Å—Ç–∞—Ä–∞–π—Ç–µ—Å—å —Å–≤–µ—Å—Ç–∏ –≤—Å—ë –∫ –º–∞—Ç—Ä–∏—á–Ω—ã–º –æ–ø–µ—Ä–∞—Ü–∏—è–º –∏–∑ `numpy` (–Ω–æ, –æ–ø—è—Ç—å –∂–µ, –Ω–µ `np.vectorize` –∏–ª–∏ —á–µ–º—É-—Ç–æ –ø–æ–¥–æ–±–Ω–æ–º—É). –ß—Ç–æ–±—ã —É–±–µ–¥–∏—Ç—å—Å—è, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å –∏–º–µ–Ω–Ω–æ —Ç–æ, —á—Ç–æ –Ω—É–∂–Ω–æ, –Ω–∞–ø–∏—à–∏—Ç–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —Ç–µ—Å—Ç—ã —Å–æ —Å–ª—É—á–∞–π–Ω—ã–º–∏ –º–∞—Ç—Ä–∏—Ü–∞–º–∏.

**[5 –±–∞–ª–ª–æ–≤]** –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `prod_and_sq_sum(A)`, –≤—ã—á–∏—Å–ª—è—é—â—É—é –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –∏ —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã `A`.

```python
def prod_and_sq_sum(matrix: np.array):
  diag = np.diag(matrix)**2
  return np.prod(diag), np.sum(diag)
```

```python
matrix = np.array([[2, 3],
       [3, 3]])
assert prod_and_sq_sum(matrix) == (36, 13)
```

```python
matrix = np.array([[9, 3, 5],
       [3, 7, 5],
       [9, 7, 2]])
assert prod_and_sq_sum(matrix) == (15876, 134)
```

```python
matrix = np.array([[3, 1, 4, 2],
       [8, 4, 2, 6],
       [6, 4, 6, 2],
       [2, 4, 8, 7]])
assert prod_and_sq_sum(matrix) == (254016, 110)
```

**[5 –±–∞–ª–ª–æ–≤]** –î–ª—è –º–∞—Ç—Ä–∏—Ü `A` –∏ `B` —Ä–∞–∑–º–µ—Ä–∞ $m\times n$ –æ–±–æ–∑–Ω–∞—á–∏–º —á–µ—Ä–µ–∑ $a_1,\ldots,a_n$ –∏ $b_1,\ldots,b_n$ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ –∏—Ö —Å—Ç–æ–ª–±—Ü—ã. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `f(A, B, k)`, –≤—ã—á–∏—Å–ª—è—é—â—É—é

$$\sum_{i=1}^{\min(k, n)}a_ib_i^T$$

```python
def fck_matrix(A: np.array, B: np.array, k: int):
    min_el = np.min((k, A.shape[1]))
    res = np.dot(A[:, :min_el+1], B[:min_el+1, :].T)
    return(res)
```

```python
A = np.array([[1, 1, 2],[3, 3, 2],[3, 3, 1], [3, 3, 1]])
B = np.array([[3, 3, 2],[2, 1, 3],[3, 2, 1],[3, 3, 1]])

fck_matrix(A,B,k=3)
```

–ù–∏–∂–Ω–∏–π –∏–Ω–¥–µ–∫—Å —É —Ä—è–¥–∞ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å i=1, –µ—Å–ª–∏ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å —Å—Ç–æ–ª–±–µ—Ü a1 –∫–∞–∫ —Å—Ç–æ–ª–±–µ—Ü –∏–Ω–¥–µ–∫—Å–∞ 0, —Ç–æ –∏ —Å—É–º–º–∏—Ä–æ–≤–∞—Ç—å –±—É–¥–µ–º –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º, –Ω–∞—á–∏–Ω–∞—è —Å –Ω—É–ª–µ–≤–æ–≥–æ

**[5 –±–∞–ª–ª–æ–≤]** –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `get_diag(A,B)`, –ø—Ä–∏–Ω–∏–º–∞—é—â—É—é –¥–≤–µ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã—Ö –º–∞—Ç—Ä–∏—Ü—ã `A` –∏ `B` –æ–¥–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –∏ –≤–æ–∑–≤—Ä–∞—â–∞—é—â—É—é –≤–µ–∫—Ç–æ—Ä –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è `AB`, –Ω–µ –≤—ã—á–∏—Å–ª—è—è –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–µ–ª–∏–∫–æ–º.

```python
def get_diag(A: np.array, B: np.array):
  return np.sum((A.T)*B, axis = 0)
```

```python
A = np.array([[3, 1, 4], [2, 1, 4], [4, 4, 1]])
B = np.array([[3, 4, 1], [2, 4, 1], [4, 1, 2]])

get_diag(A, B)
```

```python
A = np.array([[1, 1, 2], [3, 3, 2], [3, 3, 1]])
B = np.array([[3, 3, 2], [2, 1, 3], [3, 3, 1]])
get_diag(A, B)
```

–ü—É—Å—Ç—å —É –Ω–∞—Å –≤ –≤—ã–±–æ—Ä–∫–µ –µ—Å—Ç—å $10000$ –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –≤ –∫–æ—Å–º–æ–Ω–∞–≤—Ç—ã. –ò–∑ –Ω–∏—Ö —Ç–æ–ª—å–∫–æ $10$ —á–µ–ª–æ–≤–µ–∫ –ø—Ä–æ–π–¥—É—Ç –≤—Å–µ –∏—Å–ø—ã—Ç–∞–Ω–∏—è –∏ –ø—Ä–∏—Å—Ç—É–ø—è—Ç –∫ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫–∞–º. –û—Ç–±–æ—Ä –≤ –∫–æ—Å–º–æ–Ω–∞–≤—Ç—ã –Ω–∞ $95\%$ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –Ω–∞–≤—ã–∫–æ–≤ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞ –∏ –Ω–∞ $5\%$ –æ—Ç –µ–≥–æ —É–¥–∞—á–ª–∏–≤–æ—Å—Ç–∏. –ù–∞–≤—ã–∫–∏ –∏ —É–¥–∞—á–ª–∏–≤–æ—Å—Ç—å –Ω–µ –∑–∞–≤–∏—Å—è—Ç –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞.

__–∞) [5 –±–∞–ª–ª–æ–≤]__  –°–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ –Ω–∞–≤—ã–∫ –∫–∞–∂–¥–æ–≥–æ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞ –∏–∑ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ $[0; 1]$. –ü–æ –∞–Ω–∞–ª–æ–≥–∏–∏ —Å–≥–µ–Ω–∏—Ä–∏—Ä—É–π—Ç–µ —É–¥–∞—á–ª–∏–≤–æ—Å—Ç—å –∫–∞–∂–¥–æ–≥–æ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞. –ò—Ç–æ–≥–æ–≤—É—é –≤–µ–ª–∏—á–∏–Ω—É, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –±—É–¥–µ—Ç –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—å—Å—è –æ—Ç–±–æ—Ä –≤ –∫–æ—Å–º–æ–Ω–∞–≤—Ç—ã, —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –ø–æ —Ñ–æ—Ä–º—É–ª–µ:

$$
\text{score} = 0.95 \cdot \text{skill} + 0.05 \cdot \text{luck}.
$$

–ü–æ–≤—Ç–æ—Ä–∏—Ç–µ —ç—Ç—É —Å–∏–º—É–ª—è—Ü–∏—é $10^4$ —Ä–∞–∑. –û—Ü–µ–Ω–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É–¥–∞—á–ª–∏–≤–æ—Å—Ç–∏ —Ç–µ—Ö, –∫—Ç–æ —Å—Ç–∞–ª –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–º. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —É–¥–∞—á–∏ —Å—Ä–µ–¥–∏ —Ç–µ—Ö, –∫—Ç–æ —Å—Ç–∞–ª –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–º.

**Hint:** –æ–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é `np.argpartition`. –û–Ω–∞ –ø–æ–º–æ–∂–µ—Ç –≤–∞–º —Å–¥–µ–ª–∞—Ç—å –∫–æ–¥ –ø—Ä–æ—â–µ. –¢–∞–∫–∂–µ –æ–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –≤ –ø—É–Ω–∫—Ç–∞—Ö –±) –∏ –≤) –Ω—É–∂–Ω–æ –ø–µ—Ä–µ–∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–∏–º—É–ª—è—Ü–∏–π –∏–∑ –ø—É–Ω–∫—Ç–∞ –∞). –ò–Ω–∞—á–µ –∫–æ–¥ –±—É–¥–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å —Å–ª–∏—à–∫–æ–º –¥–æ–ª–≥–æ.

```python
np.random.seed(42)
n, size = 10, 10**4
skill, luck = np.random.rand(size,size), np.random.rand(size,size)
score = 0.95 * skill + 0.05 * luck

top_score = (np.argpartition(score,-n, axis = 1))[0:,-n:] # –∏—â–µ–º 10 –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã—Ö score, –≤–æ–∑–≤—Ä–∞—Ç –∏–Ω–¥–µ–∫—Å–æ–≤
res = np.take_along_axis(luck, top_score, axis = 1) #–≤–æ–∑–≤—Ä–∞—â–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è luck –ø–æ –∏–Ω–¥–µ–∫—Å–∞–º top_score
print(f'–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É–¥–∞—á–ª–∏–≤–æ—Å—Ç–∏ –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–≤ = {np.mean(res).round(2)}') # —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç –≤—Å–µ—Ö –≤–µ–∑—É–Ω—á–∏–∫–æ–≤-–∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–≤
```

__–±) [5 –±–∞–ª–ª–æ–≤]__  –û—Ç–±–µ—Ä–∏—Ç–µ –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–≤ –ø–æ –Ω–∞–≤—ã–∫–∞–º, `skill`. –û—Ü–µ–Ω–∏—Ç–µ —Å –ø–æ–º–æ—â—å—é —Å–∏–º—É–ª—è—Ü–∏–π, –Ω–∞ —Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–ª–æ–≤–µ–∫ —Ç–æ–ø –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –ø–æ `skill` –ø–µ—Ä–µ—Å–µ–∫–∞–µ—Ç—Å—è —Å —Ç–æ–ø–æ–º –ø–æ `score`.

```python
top_skill = (np.argpartition(skill,-n, axis = 1))[0:,-n:]
intersection = [len(np.intersect1d(top_skill[i], top_score[i])) for i in range(skill.shape[0])]
print(f'–°—Ä–µ–¥–Ω—è—è –¥–æ–ª—è –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–π = {round(np.mean(np.array(intersection))/10, 2)}')
```

__–≤) [5 –±–∞–ª–ª–æ–≤]__ –ê —á—Ç–æ –±—É–¥–µ—Ç, –µ—Å–ª–∏ –º—ã –±—É–¥–µ–º –æ—Ç–±–∏—Ä–∞—Ç—å –±–æ–ª—å—à–µ –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–≤?

–î–ª—è –∫–∞–∂–¥–æ–≥–æ $n$ –æ—Ç $10$ –¥–æ $500$ —Å —à–∞–≥–æ–º –≤ $10$ —Å–¥–µ–ª–∞–π—Ç–µ –ø—Ä–æ—Ü–µ–¥—É—Ä—É –æ—Ç–±–æ—Ä–∞ –≤ –∫–æ—Å–º–æ–Ω–∞–≤—Ç—ã —Ç–æ–ø-$n$ –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –ø–æ `skill` –∏ –ø–æ `score`. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ, –Ω–∞ —Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç–æ–≤ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è —ç—Ç–∏ —Ç–æ–ø—ã. –ò–∑–æ–±—Ä–∞–∑–∏—Ç–µ –ø–æ–ª—É—á–∏–≤—à—É—é—Å—å—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ. –ü–æ –æ—Å–∏ –∞–±—Å—Ü–∏—Å—Å –æ—Ç–ª–æ–∂–∏—Ç–µ $n$, –ø–æ –æ—Å–∏ –æ—Ä–¥–∏–Ω–∞—Ç –ø—Ä–æ—Ü–µ–Ω—Ç –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è.

!!! –û–°–¢–û–†–û–ñ–ù–û –†–ê–ë–û–¢–ê–ï–¢ +- 3 –ú–ò–ù–£–¢–´

```python
def top_top(n):
  top_skill = (np.argpartition(skill,-n, axis = 1))[0:,-n:] # –æ—Ç–±–∏—Ä–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ n —Ç–æ–ø–æ–≤ –ø–æ —Å–∫–∏–ª–ª—É
  top_score = (np.argpartition(score,-n, axis = 1))[0:,-n:] # –æ—Ç–±–∏—Ä–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ n —Ç–æ–ø–æ–≤ –ø–æ —Å–∫–æ—Ä—É
  intersection = [len(np.intersect1d(top_skill[i], top_score[i])) for i in range(skill.shape[0])] # –∏—â–µ–º –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ –ø–æ –∏–Ω–¥–µ–∫—Å–∞–º –ø–æ —Å—Ç—Ä–æ–∫–∞–º –¥–≤—É—Ö –º–∞—Ç—Ä–∏—Ü
  return np.mean(np.array(intersection)/n) # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Å—Ä–µ–¥–Ω—é—é –¥–æ–ª—é –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–π
```

```python
%%time
mass = list()
for n in range(10,510,10):
  mass.append(top_top(n))
```

```python
plt.plot(np.arange(10,510,10), mass)
plt.title("–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ —Ç–æ–ø n –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –ø–æ skill –∏ –ø–æ score");
plt.ylabel('–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å');
plt.xlabel('N-–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤');
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5);
```

__–≥) [5 –±–∞–ª–ª–æ–≤]__ –°–¥–µ–ª–∞–π—Ç–µ –≤—ã–≤–æ–¥—ã –∏–∑ –ø—Ä–æ–¥–µ–ª–∞–Ω–Ω—ã—Ö —Ä–∞—Å—Å—á—ë—Ç–æ–≤.

__–û—Ç–≤–µ—Ç:__ –° —É–≤–µ–ª–∏—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ n, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–π –ø–æ —Å—Ç—Ä–æ–∫–∞–º –º–∞—Ç—Ä–∏—Ü score –∏ skill —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —á—Ç–æ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–æ–ø–æ–≤ —Å–æ–≤–ø–∞–¥—É—Ç –±—É–¥–µ—Ç –±–æ–ª—å—à–µ, –ø—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Ç–æ–±—Ä–∞–Ω–Ω—ã—Ö –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–≤. –°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –µ—Å–ª–∏ –º—ã –∏—â–µ–º –æ–±—Ä–∞—Ç–Ω–æ–µ, —Ç–æ –µ—Å—Ç—å –æ—Ç—Ä–∏—Ü–∞–Ω–∏–µ (1- –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ), —Ç–æ —ç—Ç–∞ –≤–µ–ª–∏—á–∏–Ω–∞ –±—É–¥–µ—Ç —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è —Å —Ä–æ—Å—Ç–æ–º n.

> –ü–æ–≤–æ–¥ –ª–∏ —ç—Ç–æ –≥—Ä—É—Å—Ç–∏—Ç—å? –ö–æ–Ω–µ—á–Ω–æ –∂–µ, –Ω–µ—Ç. –ï—Å–ª–∏ –≤—ã –ø–æ–ø—ã—Ç–∞–ª–∏—Å—å —Å—Ö–≤–∞—Ç–∏—Ç—å—Å—è –∑–∞ –æ–¥–Ω—É –∏–∑ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–µ–π –∏ —É –≤–∞—Å –Ω–∏—á–µ–≥–æ –Ω–µ –≤—ã—à–ª–æ, –∑–∞–±—É–¥—å—Ç–µ –æ–± —ç—Ç–æ–º. –ù–µ –≤–ø–∞–¥–∞–π—Ç–µ –≤ –¥–µ–ø—Ä–µ—Å—Å–∏—é, –ø—Ä–æ–¥–æ–ª–∂–∞–π—Ç–µ –ø—ã—Ç–∞—Ç—å—Å—è. –í –º–∏—Ä–µ –µ—Å—Ç—å —Å—Ç–æ–ª—å–∫–æ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ —É–ø—É—Å—Ç–∏—Ç—å. –û–±–∏–¥–Ω–æ –±—É–¥–µ—Ç, –µ—Å–ª–∏ –≤—ã –Ω–µ –ø–æ–ø—Ä–æ–±—É–µ—Ç–µ –µ—â—ë. –†–∞–Ω–æ –∏–ª–∏ –ø–æ–∑–¥–Ω–æ —É–¥–∞—á–∞ –≤–∞–º —É–ª—ã–±–Ω—ë—Ç—Å—è ‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è

–î–∞–≤–∞–π—Ç–µ –ø–æ–ø—Ä–æ–±—É–µ–º –ø–æ–Ω—è—Ç—å, –∫–æ–≥–¥–∞ –∏–º–µ–Ω–Ω–æ –æ–Ω–∞ –≤–∞–º —É–ª—ã–±–Ω—ë—Ç—Å—è.

__–¥) [4 –±–∞–ª–ª–∞]__ –Æ—Ä–∞ —Ö–æ—á–µ—Ç —Å—Ç–∞—Ç—å –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–º. –£ –Ω–µ–≥–æ –¥–æ–≤–æ–ª—å–Ω–æ —Ö–æ—Ä–æ—à–æ —Ä–∞–∑–≤–∏—Ç—ã –Ω–∞–≤—ã–∫–∏. –Æ—Ä–∞ –≤—Ö–æ–¥–∏—Ç –≤ $1\%$ –ª—É—á—à–∏—Ö, –µ–≥–æ `score = 0.99` (–Æ—Ä–∞ –ø—Ä–æ–∫–∞—á–∞–ª —Å–≤–æ–∏ –Ω–∞–≤—ã–∫–∏ –¥–æ 0.99 –∏ –≤–æ–æ–±—â–µ –Ω–µ –ø–æ–ª–∞–≥–∞–µ—Ç—Å—è –Ω–∞ —É–¥–∞—á—É). –° –∫–∞–∫–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –Æ—Ä–∞ —Å—Ç–∞–Ω–µ—Ç –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–º?

```python
top_score.shape # —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –º–∞—Ç—Ä–∏—Ü—ã —Ç–æ–ø 10 –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –ø–æ —Å–∫–æ—Ä—É
```

```python
top_score_values = np.take_along_axis(score, top_score, axis = 1) # top_score –ø–æ–ª—É—á–∏–ª–∏ –≤ –ø—É–Ω–∫—Ç–µ –ê –≤ –≤–∏–¥–µ –∏–Ω–¥–µ–∫—Å–æ–≤
# –ø–æ –∏–Ω–¥–µ–∫—Å–∞–º top_score –ø–æ–ª—É—á–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ score, –∑–∞–Ω–µ—Å–µ–º –≤—Å–µ –≤ –æ—Ç–¥–µ–ª—å–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É top_score_values, —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å (10000, 10)

great_yura = np.apply_along_axis(lambda row: min(row)<0.99, axis=1, arr = top_score_values)
# –∏—â–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ top_score_values –ø–æ —Å—Ç—Ä–æ–∫–∞–º, –µ—Å–ª–∏ –æ–Ω–æ –º–µ–Ω—å—à–µ —á–µ–º 0.99 (—Å–∫–æ—Ä –Æ—Ä—ã), —Ç–æ –±—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ –≤ —ç—Å–∫–ø–µ—Ä–∏–º–µ–Ω—Ç–µ –Æ—Ä–∞ –æ–±–≥–æ–Ω–∏—Ç
# —É—á–∞—Å—Ç–Ω–∏–∫–∞ —Å –º–µ–Ω—å—à–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º –∏ —Å—Ç–∞–Ω–µ—Ç –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–º
print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å—Ç–∞—Ç—å –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–º –¥–ª—è –Æ—Ä—ã = {np.mean(great_yura).round(2)}')
```

__–≤) [5 –±–∞–ª–ª–æ–≤]__  –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –≤—Ä–µ–º–µ–Ω–∏ –æ–∂–∏–¥–∞–Ω–∏—è –ª–∏—Ñ—Ç–∞ –¥–ª—è –†–∞–π–∞–Ω–∞ –ì–æ—Å–ª–∏–Ω–≥–∞ –∏ –≥—Ä—É–ø–ø—ã Radiohead. –ü–æ—Ö–æ–∂–∏ –ª–∏ –≤–∏–∑—É–∞–ª—å–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã–µ? –ù–∞—Å–∫–æ–ª—å–∫–æ —Å–∏–ª—å–Ω–æ –æ–Ω–∏ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞?

–ù–∞–¥–æ –±—É–¥–µ—Ç –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –ª–∏—Ñ—Ç –¥–≤–∏–∂–µ—Ç—Å—è –ø–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—é –∏–ª–∏ –ø—Ä–æ—Ç–∏–≤ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è –∫ –ø–µ—Ä—Å–æ–Ω–∞–∂—É, –±—É–¥–µ–º –æ—Ç—Ç–∞–ª–∫–∏–≤–∞—Ç—å—Å—è –æ—Ç —ç—Ç–æ–≥–æ –ø—Ä–∏ –ø–æ–¥—Å—á–µ—Ç–µ –≤—Ä–µ–º–µ–Ω–∏ –æ–∂–∏–¥–∞–Ω–∏—è. –ï—Å–ª–∏ –ì–æ—Å–ª–∏–Ω–≥ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ 3 —ç—Ç–∞–∂–µ, –∞ –ª–∏—Ñ—Ç —É—Å–ª–æ–≤–Ω–æ –Ω–∞ 2-–æ–º, —Ç–æ –ø—Ä–∏ –¥–≤–∏–∂–µ–Ω–∏–∏ –ª–∏—Ñ—Ç–∞ –≤–≤–µ—Ä—Ö (–ø–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—é) –≥–µ—Ä–æ–π –î—Ä–∞–π–≤–∞ –±—É–¥–µ—Ç –∂–¥–∞—Ç—å 1 –º–∏–Ω—É—Ç—É, –∞ –µ—Å–ª–∏ –ª–∏—Ñ—Ç –¥–≤–∏–∂–µ—Ç—Å—è –≤–Ω–∏–∑ –∫ 0 —ç—Ç–∞–∂—É –∏ –ø–æ—Ç–æ–º –≤–≤–µ—Ä—Ö (–ø—Ä–æ—Ç–∏–≤ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è), —Ç–æ 2 + 3 = 5 –º–∏–Ω—É—Ç –æ–∂–∏–¥–∞–Ω–∏—è. –°–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤–µ–∫—Ç–æ—Ä vers –∏–∑ 0 –∏ 1, –≥–¥–µ 1 - –¥–≤–∏–∂–µ–Ω–∏–µ –ø–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—é, –∞ 0 - –¥–≤–∏–∂–µ–Ω–∏–µ –ª–∏—Ñ—Ç–∞ —Å —Ä–∞–∑–≤–æ—Ä–æ—Ç–æ–º

```python
np.random.seed(2)
vers, stops  = np.random.randint(2, size=(10**4)), np.random.uniform(0, 30, (10**4))
# –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤–µ–∫—Ç–æ—Ä vers —Å –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è–º–∏ –¥–≤–∏–∂–µ–Ω–∏—è –ª–∏—Ñ—Ç–∞
# –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤–µ–∫—Ç–æ—Ä stops —Å –Ω–∞—á–∞–ª—å–Ω–æ–π –ø–æ–∑–∏—Ü–∏–µ–π –ª–∏—Ñ—Ç–∞

def waiting(n):
  time_straight = abs(stops[np.where(vers == 1)[0]]-n) # –ª–∏—Ñ—Ç –¥–≤–∏–∂–µ—Ç—Å—è –ø–æ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—é –∫ –ø–∞—Å—Å–∞–∂–∏—Ä—É (–ø—Ä–æ—Å—Ç–æ –≤–≤–µ—Ä—Ö –∏–ª–∏ –≤–Ω–∏–∑, –ø–æ—ç—Ç–æ–º—É –º–æ–¥—É–ª—å –≤–æ–∑—å–º—É)
  time_reverse = stops[np.where(vers == 0)[0]] # —Ç–∞–º –≥–¥–µ vers == 0, –ª–∏—Ñ—Ç –¥–≤–∏–∂–µ—Ç—Å—è —Å —Ä–∞–∑–≤–æ—Ä–æ—Ç–æ–º, –ø–æ–ª—É—á–∞–µ–º –ø–æ —Å–æ–æ—Ç–≤–µ—Ç—Å–≤—É—é—â–∏–º –∏–Ω–¥–µ–∫—Å–∞–º —ç—Ç–∞–∂–∏ –≤ stops
  time_reverse = np.where(time_reverse>n, 60 - time_reverse - n, time_reverse + n)
  # —Ç–∞–º –≥–¥–µ —ç—Ç–∞–∂ –±–æ–ª—å—à–µ —ç—Ç–∞–∂–∞ n^
  return np.concatenate([time_straight, time_reverse])
```

- –í –ø–µ—Ä–≤–æ–º –ø–æ –æ—Å–∏ –∞–±—Å—Ü–∏—Å—Å –æ—Ç–ª–æ–∂–∏—Ç–µ —ç—Ç–∞–∂, —Å –∫–æ—Ç–æ—Ä–æ–≥–æ —Å—Ç–∞—Ä—Ç—É–µ—Ç –°—Ç–∏–≤, –∞ –ø–æ –æ—Å–∏ –æ—Ä–¥–∏–Ω–∞—Ç –æ—Ç–ª–æ–∂–∏—Ç–µ –æ—Ü–µ–Ω–∫—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ç–æ–≥–æ, —á—Ç–æ –ª–∏—Ñ—Ç –ø—Ä–∏–¥—ë—Ç —Å–≤–µ—Ä—Ö—É.
- –í –ø–µ—Ä–≤–æ–º –ø–æ –æ—Å–∏ –∞–±—Å—Ü–∏—Å—Å –æ—Ç–ª–æ–∂–∏—Ç–µ —ç—Ç–∞–∂, —Å –∫–æ—Ç–æ—Ä–æ–≥–æ —Å—Ç–∞—Ä—Ç—É–µ—Ç –°—Ç–∏–≤, –∞ –ø–æ –æ—Å–∏ –æ—Ä–¥–∏–Ω–∞—Ç —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –æ–∂–∏–¥–∞–Ω–∏—è –ª–∏—Ñ—Ç–∞.

Waiting –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç—Å—è 10k –∑–Ω–∞—á–µ–Ω–∏–π –≤—Ä–µ–º–µ–Ω–∏ –æ–∂–∏–¥–∞–Ω–∏—è, —É—Å—Ä–µ–¥–Ω—è–µ–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —ç—Ç–∞–∂–∞ —ç—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏—è

–ö–∞–∫ –¥—É–º–∞–µ—Ç–µ, –±—É–¥–µ—Ç –ª–∏ –Ω–∞–±–ª—é–¥–∞—Ç—å—Å—è —á—Ç–æ-—Ç–æ –ø–æ—Ö–æ–∂–µ–µ –≤ —Ä–µ–∞–ª—å–Ω–æ–π –∂–∏–∑–Ω–∏? –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –ø–æ–Ω–∞–±–ª—é–¥–∞—Ç—å –∑–∞ –ª–∏—Ñ—Ç–æ–º —É —Å–µ–±—è –¥–æ–º–∞ –∏ –≤ –æ—Ñ–∏—Å–µ. –ù–∞ –∫–∞–∫–æ–º —ç—Ç–∞–∂–µ –≤—ã –∂–∏–≤—ë—Ç–µ/—Ä–∞–±–æ—Ç–∞–µ—Ç–µ? –í—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –ª–∏ –¥–ª—è –≤–∞—Å [–ø–∞—Ä–∞–¥–æ–∫—Å –ª–∏—Ñ—Ç–∞?](https://en.wikipedia.org/wiki/Elevator_paradox)

## –ó–∞–¥–∞—á–∞ 4: –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∏ (30 –±–∞–ª–ª–æ–≤)

–û—Ç —Å—Ç–∞–Ω—Ü–∏–∏ –†–æ–º–∞—à–∫–æ–≤–æ –¥—Ä—É–≥ –∑–∞ –¥—Ä—É–≥–æ–º –≤ –æ–¥–Ω–æ–º –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–∏ –æ—Ç—Ö–æ–¥–∏—Ç $100$ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–æ–≤ —Å –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–º –≤ –æ–¥–Ω—É –º–∏–Ω—É—Ç—É. –°—Ç–∞—Ä—Ç–æ–≤—ã–µ —Å–∫–æ—Ä–æ—Å—Ç–∏ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–æ–≤ -- –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ, —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω—ã–µ –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ $[0;1]$ —Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã. –ö–æ–≥–¥–∞ –±–æ–ª–µ–µ –±—ã—Å—Ç—Ä—ã–π –ø–∞—Ä–æ–≤–æ–∑–∏–∫ –¥–æ–≥–æ–Ω—è–µ—Ç –±–æ–ª–µ–µ –º–µ–¥–ª–µ–Ω–Ω—ã–π, –æ–Ω —Å–Ω–∏–∂–∞–µ—Ç —Å–≤–æ—é —Å–∫–æ—Ä–æ—Å—Ç—å –¥–æ —Å–∫–æ—Ä–æ—Å—Ç–∏ –¥–æ–≥–Ω–∞–Ω–Ω–æ–≥–æ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∞.

–ß–µ—Ä–µ–∑ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–ª–∏–Ω–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –≤—Ä–µ–º–µ–Ω–∏ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∏ —Ä–∞–∑–æ–±—å—é—Ç—Å—è –Ω–∞ –Ω–µ—Å–∫–æ–ª—å–∫–æ –≥—Ä—É–ø–ø –∏ –¥–∞–ª–µ–µ —á–∏—Å–ª–æ –≥—Ä—É–ø–ø –º–µ–Ω—è—Ç—å—Å—è –Ω–µ –±—É–¥–µ—Ç. –ù–∞–∑–æ–≤—ë–º —ç—Ç–æ —á–∏—Å–ª–æ –≥—Ä—É–ø–ø —Ñ–∏–Ω–∞–ª—å–Ω—ã–º. –ü—Ä–æ–≤–µ–¥–∏—Ç–µ $1000$ —Å–∏–º—É–ª—è—Ü–∏–π –∏ –Ω–∞–π–¥–∏—Ç–µ –æ—Ç–≤–µ—Ç—ã –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–µ –≤–æ–ø—Ä–æ—Å—ã:

__–∞) [5 –±–∞–ª–ª–æ–≤]__  –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ø—è—Ç—ã–π –ø–∞—Ä–æ–≤–æ–∑–∏–∫ -- –ª–∏–¥–µ—Ä –æ—á–µ—Ä–µ–¥–Ω–æ–π –≥—Ä—É–ø–ø—ã?

```python
np.random.seed(42)
trains_speed = np.random.rand(1000,100) # —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–ª–∏ –º–∞—Ç—Ä–∏—Ü—É –Ω–∞ 1–∫ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤, 100 —Ç—Ä–∞–º–≤–∞–µ–≤
np.mean(np.argmin(trains_speed[:,0:5], axis = 1) == 4) # –µ—Å–ª–∏ –ø—è—Ç—ã–π –ø–∞—Ä–æ–≤–æ–∑ —Å–∞–º—ã–π –º–µ–¥–ª–µ–Ω–Ω—ã–π –∏–∑ 5 —Ç–æ –æ–Ω –±—É–¥–µ—Ç –ª–∏–¥–µ—Ä–æ–º –Ω–æ–≤–æ–π –≥—Ä—É–ø–ø—ã,
# –∏–Ω–∞—á–µ –ø—è—Ç—ã–π –ø–∞—Ä–æ–≤–æ–∑ (—É –Ω–µ–≥–æ –∏–Ω–¥–µ–∫—Å = 4) –¥–æ–≥–æ–Ω–∏—Ç –≤–ø–µ—Ä–µ–¥–∏—Å—Ç–æ—è—â–∏–π –±–æ–ª–µ–µ –º–µ–¥–ª–µ–Ω–Ω—ã–π –∏ –ø—Ä–∏–º–∫–Ω–µ—Ç –∫ –µ–≥–æ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–µ
```

__–±) [5 –±–∞–ª–ª–æ–≤]__ –ù–∞ —Å–∫–æ–ª—å–∫–æ —Ñ–∏–Ω–∞–ª—å–Ω—ã—Ö –≥—Ä—É–ø–ø –≤ —Å—Ä–µ–¥–Ω–µ–º —Ä–∞–∑–±–∏–≤–∞—é—Ç—Å—è –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∏?

```python
def accumulate_minimum(row):
    return len(np.unique(np.minimum.accumulate(row))) # –ø—Ä–∏–º–µ–Ω—è–µ–º –≤—Å—Ç—Ä–æ–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é –∏ —Å–º–æ—Ç—Ä–∏–º —Å–∫–æ–ª—å–∫–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –º–∞—Å—Å–∏–≤–µ
res = [accumulate_minimum(row) for row in trains_speed] # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ —Å—Ç—Ä–æ–∫–∞–º –º–∞—Ç—Ä–∏—Ü—ã, –ø—Ä–∏–º–µ–Ω—è—è –¥–∞–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é
np.mean(res)
```

!!! –í–ê–ñ–ù–û: –ï—Å–ª–∏ –∑–∞—Ö–æ—Ç–∏—Ç–µ –ø–æ–∫—Ä—É—Ç–∏—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –ø–∞—Ä—É —Ä–∞–∑, —Ç–æ –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –æ–±–Ω–æ–≤–∏—Ç–µ —Ç–∞–±–ª–∏—á–∫—É trains_speed, –Ω–∞–∂–∞–≤ –Ω–∞ –±–ª–æ–∫ –Ω–∏–∂–µ.

–í –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–µ –±—É–¥–µ–º –ø–æ –æ–¥–Ω–æ–º—É –≤—ã–ø—É—Å–∫–∞—Ç—å —Ç—Ä–∞–º–≤–∞–∏ –∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞—Ç—å –∏—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –ø—Ä–æ–π–¥–µ–Ω–Ω–æ–µ –∑–∞ –∫–∞–∂–¥—É—é –º–∏–Ω—É—Ç—É. –í —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ –¥–∏—Å—Ç–∞–Ω—Ü–∏—è –±–æ–ª–µ–µ –±—ã—Å—Ç—Ä–æ–≥–æ —Ç—Ä–∞–º–≤–∞—è –≤ –∫–∞–∫–æ–π-—Ç–æ –º–æ–º–µ–Ω—Ç –ø—Ä–µ–≤—ã—à–∞–µ—Ç –¥–∏—Å—Ç–∞–Ω—Ü–∏—é –≤–ø–µ—Ä–µ–¥–∏—Å—Ç–æ—è—â–µ–≥–æ –±–æ–ª–µ–µ –º–µ–¥–ª–µ–Ω–Ω–æ–≥–æ, —Ç–æ —Å–∫–æ—Ä–æ—Å—Ç—å –∏ –ø—Ä–æ–π–¥–µ–Ω–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –±–æ–ª–µ–µ –±—ã—Å—Ç—Ä–æ–≥–æ –∑–∞–º–µ–Ω—è–µ–º –Ω–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –±–æ–ª–µ–µ –º–µ–¥–ª–µ–Ω–Ω–æ–≥–æ

```python
np.random.seed(42)
trains_speed = np.random.rand(1000,100)
```

```python
def len_groups(trains):
  dist = np.array([])                # –ø—É—Å—Ç–æ–π –º–∞—Å—Å–∏–≤ –¥–ª—è —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π
  mass = list()                      # –∑–¥–µ—Å—å —Å—á–∏—Ç–∞–µ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å–∫–æ—Ä–æ—Å—Ç–∏ = —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –≥—Ä—É–ø–ø—ã
  for i in range(1,101):             # –ø–æ –æ—á–µ—Ä–µ–¥–∏ –≤—ã–ø—É—Å–∫–∞–µ–º –ø–æ–µ–∑–¥–∞
    dist = np.append(dist,0)+trains[:i]           # –Ω–∞–¥–æ –¥–æ–±–∞–≤–∏—Ç—å —Å–∫–æ—Ä–æ—Å—Ç—å –Ω–æ–≤–æ–≥–æ –ø–æ–µ–∑–¥–∞ –≤ –º–∞—Å—Å–∏–≤ —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è–º–∏ (—Å–∫–æ—Ä–æ—Å—Ç—å = —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –∑–∞ 1 –º–∏–Ω—É—Ç—É)
                                     # –∏–∑-–∑–∞ —Ä–∞–∑–Ω–æ–π —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –º–∞—Å—Å–∏–≤–æ–≤, –¥–æ–±–∞–≤–ª—è–µ–º —á–∏—Å–ª–æ 0 –≤ –º–∞—Å—Å–∏–≤ —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º, —á—Ç–æ–±—ã –≤—Å–µ –æ–∫ —Å–ª–æ–∂–∏–ª–æ—Å—å —Å trains[:i]
    for j in range(1,len(dist)):
      if dist[j] > dist[j - 1]:      # –∑–¥–µ—Å—å –∏—â–µ–º –ø–æ–µ–∑–¥–∞ —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º –±–æ–ª—å—à–µ, —á–µ–º —É  –≤–ø–µ—Ä–µ–¥–∏—Å—Ç–æ—è—â–µ–≥–æ –ø–æ–µ–∑–¥–∞ —Å –º–µ–Ω—å—à–µ–π —Å–∫–æ—Ä–æ—Å—Ç—å—é
        dist[j] = dist[j - 1]        # –∑–∞–º–µ–Ω—è–µ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ —É –±–æ–ª–µ–µ –±—ã—Å—Ç—Ä–æ–≥–æ –Ω–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –±–æ–ª–µ–µ –º–µ–¥–ª–µ–Ω–Ω–æ–≥–æ
        trains[:i][j] = trains[:i][j - 1]         # –∑–∞–º–µ–Ω—è–µ–º —Å–∫–æ—Ä–æ—Å—Ç–∏
    mass.append(len(np.unique(trains[:i])))       # –ø–æ—Å–ª–µ –≤—Å–µ—Ö –º–∞–Ω–∏–ø—É–ª—è—Ü–∏–π –∑–∞ 1 –º–∏–Ω—É—Ç—É —Å—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –≥—Ä—É–ø–ø
  return mass
```

```python
res = [len_groups(trains) for trains in trains_speed] # –ø–æ–≤—Ç–æ—Ä—è–µ–º –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ –≤ –º–∞—Ç—Ä–∏—Ü–µ (100 –ø–æ–µ–∑–¥–æ–≤ —Å —Å—Ç—Ä–æ–∫–µ)
# –í res —Ö—Ä–∞–Ω–∏—Ç—Å—è 1000 —Å—Ç—Ä–æ–∫ (—Å–∏–º—É–ª—è—Ü–∏–π) –ø–æ 100 —Å—Ç–æ–ª–±—Ü–æ–≤ (—É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –≥—Ä—É–ø–ø –∑–∞ –∫–∞–∂–¥—É—é –º–∏–Ω—É—Ç—É, –≤—Å–µ–≥–æ 100 –º–∏–Ω—É—Ç)
```

–í—ã—à–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∞ —Å–∏—Ç—É–∞—Ü–∏—è –Ω–∞ –≥–æ—Ä–∏–∑–æ–Ω—Ç–µ 100 –º–∏–Ω—É—Ç. –ü–æ—Å–º–æ—Ç—Ä–∏–º, —á—Ç–æ –±—É–¥–µ—Ç, –µ—Å–ª–∏ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –ø—Ä–æ–º–µ–∂—É—Ç–æ–∫ –¥–æ 200 –º–∏–Ω—É—Ç, –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏ –∫ —Ä–µ—à–µ–Ω–∏—é –±—É–¥—É—Ç –≤ –ø—É–Ω–∫—Ç–µ –ì, —è —Å–Ω–∞—á–∞–ª–∞ –ì —Å–¥–µ–ª–∞–ª–∞, –ø–æ—Ç–æ–º –º–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–ª–∞ –í –¥–æ  250 –º–∏–Ω—É—Ç


–û–ë–Ø–ó–ê–¢–ï–õ–¨–ù–û –ù–£–ñ–ù–û –û–ë–ù–û–í–ò–¢–¨ trains_speed

```python
np.random.seed(52)
trains_speed = np.random.rand(1000,100)
```

```python
def len_groups(trains):
  dist = np.array([])
  mass = list()
  for i in range(1,251):
    if len(dist)<100:
      dist = np.append(dist,0)+trains[:i]
    else:
      dist = dist+trains[:i]
    for j in range(1,len(dist)):
      if dist[j] > dist[j - 1]:
        dist[j] = dist[j - 1]
        trains[:i][j] = trains[:i][j - 1]
    mass.append(len(np.unique(trains[:i])))
  return mass
```

```python
res = [len_groups(trains) for trains in trains_speed]
```

```python
res = np.mean(res,axis = 0)
```

–ó–∞—á–µ–º –∑–¥–µ—Å—å –∞–∂ –¥–≤–æ–π–Ω–æ–π —Ü–∏–∫–ª –≤ —Ñ—É–Ω–∫—Ü–∏–∏? –Ø —á–µ—Å—Ç–Ω–æ –ø—ã—Ç–∞–ª–∞—Å—å –æ–±–æ–π—Ç–∏—Å—å —Ç–æ–ª—å–∫–æ –æ–¥–Ω–∏–º —Ü–∏–∫–ª–æ–º –∏ –ø–æ—Ç–æ–º –ø—Ä–∏–º–µ–Ω—è—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –∫ —Å—Ç—Ä–æ–∫–∞–º –º–∞—Ç—Ä–∏—Ü—ã trains_speed, –Ω–æ –µ—Å—Ç—å —É–∑–∫–∏–π –º–æ–º–µ–Ω—Ç, –∫–æ—Ç–æ—Ä—ã–π —è –Ω–µ —Å–º–æ–≥–ª–∞ –æ–±–æ–π—Ç–∏ –±–µ–∑ –≤—Ç–æ—Ä–æ–≥–æ —Ü–∏–∫–ª–∞. –ú–æ–º–µ–Ω—Ç –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ –≤ –æ–¥–Ω—É –º–∏–Ω—É—Ç—É —Å—Ä–∞–∑—É –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–æ–µ–∑–¥–æ–≤ –º–æ–≥—É—Ç –¥–æ–≥–Ω–∞—Ç—å –±–æ–ª–µ–µ –º–µ–¥–ª–µ–Ω–Ω—ã–π –≤–ø–µ—Ä–µ–¥–∏—Å—Ç–æ—è—â–∏–π. –¢–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –≤ —ç—Ç—É –∂–µ –º–∏–Ω—É—Ç—É –º–µ–Ω—è—Ç—å —Å–∫–æ—Ä–æ—Å—Ç–∏ —É –≤—Å–µ—Ö –±–æ–ª–µ–µ –±—ã—Å—Ç—Ä—ã—Ö, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∏ —É–∂–µ –æ–±—Ä–∞–∑–æ–≤–∞–ª–∏ –≥—Ä—É–ø–ø—É.

–ü—Ä–æ—Å—Ç–æ–µ –ø—Ä–∏—Å–≤–æ–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—é —Å–ø—Ä–∞–≤–∞ (–±–æ–ª–µ–µ –±—ã—Å—Ç—Ä—ã–π) –∑–Ω–∞—á–µ–Ω–∏–µ —Å–ª–µ–≤–∞ (–±–æ–ª–µ–µ –º–µ–¥–ª–µ–Ω–Ω—ã–π) –±–µ–∑ —Ü–∏–∫–ª–∞ –±—É–¥–µ—Ç –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ, —Ç–∞–∫ –∫–∞–∫ –Ω–µ –ø–æ–ª—É—á–∏—Ç—Å—è –∑–∞ –æ–¥–Ω—É –º–∏–Ω—É—Ç—É –ø–æ–º–µ–Ω—è—Ç—å –æ–±–µ —Å–∫–æ—Ä–æ—Å—Ç–∏ –Ω–∞ –æ–±—â—É—é –±–æ–ª–µ–µ –º–µ–¥–ª–µ–Ω–Ω—É—é —Å—Ä–∞–∑—É.

–ö–∞–∫ –ø—Ä–∞–≤–∏–ª–æ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∏ –Ω–µ –º–æ–≥—É—Ç —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å "–∏–¥–µ–∞–ª—å–Ω—ã–π —Ä—è–¥" –∑–∞ 100 –º–∏–Ω—É—Ç, –±–æ–ª–µ–µ –±—ã—Å—Ç—Ä—ã–º –∏–Ω–¥–∏–≤–∏–¥–∞–º, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã–µ—Ö–∞–ª–∏ –ø–æ–∑–¥–Ω–æ —Ç—Ä–µ–±—É–µ—Ç—Å—è –±–æ–ª—å—à–µ –≤—Ä–µ–º–µ–Ω–∏, —á—Ç–æ–±—ã –ø—Ä–æ–µ—Ö–∞—Ç—å —Ä–∞–∑–Ω–∏—Ü—É –≤ –¥–∏—Å—Ç–∞–Ω—Ü–∏–∏ —Å –±–æ–ª–µ–µ –º–µ–¥–ª–µ–Ω–Ω—ã–º –ø–æ–µ–∑–¥–æ–º, –∞ —Ç–æ–ª—å–∫–æ –ø—Ä–∏ —ç—Ç–æ–º —É—Å–ª–æ–≤–∏–∏ –º–æ–∂–µ—Ç –ø—Ä–æ–∏–∑–æ–π—Ç–∏ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ —Å–æ —Å–º–µ–Ω–æ–π —Å–∫—Ä–æ—Å—Ç–µ–π. –í—Å–µ —ç—Ç–æ –≤—Ä–µ–º—è –æ–Ω–∏ –±—É–¥—É—Ç –µ—Ö–∞—Ç—å –≤–¥–∞–ª–∏ —Å–æ —Å–≤–æ–µ–π —É–Ω–∏–∫–∞–ª—å–Ω–æ–π —Å–∫–æ—Ä–æ—Å—Ç—å—é –∏ —Å—á–∏—Ç–∞—Ç—å—Å—è –∫–∞–∫ —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –≥—Ä—É–ø–ø—ã.

–°–æ –≤—Ä–µ–º–µ–Ω–µ–º –æ–Ω–∏ —Ñ–æ—Ä–º–∏—Ä—É—é—Ç —Ç–æ—Ç —Å–∞–º—ã–π "–∏–¥–µ–ª–∞—å–Ω—ã–π —Ä—è–¥".

__–≥) [10 –±–∞–ª–ª–æ–≤]__  –û—Ü–µ–Ω–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è, –≤ —Ç–µ—á–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∏ —Ä–∞–∑–±–∏–≤–∞—é—Ç—Å—è –Ω–∞ —Ñ–∏–Ω–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –≥—Ä—É–ø–ø. –û—Ü–µ–Ω–∏—Ç–µ –¥–ª—è —ç—Ç–æ–π –∂–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –¥–∏—Å–ø–µ—Ä—Å–∏—é.

!!!–û–°–¢–û–†–û–ñ–ù–û –†–ê–ë–û–¢–ê–ï–¢ 20 –°–ï–ö–£–ù–î

!!! –í–ê–ñ–ù–û: –ï—Å–ª–∏ –∑–∞—Ö–æ—Ç–∏—Ç–µ –ø–æ–∫—Ä—É—Ç–∏—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –ø–∞—Ä—É —Ä–∞–∑, —Ç–æ –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –æ–±–Ω–æ–≤–∏—Ç–µ —Ç–∞–±–ª–∏—á–∫—É trains_speed, –Ω–∞–∂–∞–≤ –Ω–∞ –±–ª–æ–∫ –Ω–∏–∂–µ.

```python
np.random.seed(42)
trains_speed = np.random.rand(1000,100)
```

```python
def final_groups_time(trains):
  i = 0
  dist = np.array([])
  while len(np.unique(trains)) != len(np.unique(np.minimum.accumulate(trains))):      # –±—É–¥–µ–º –≤—ã–ø–æ–ª–Ω—è—Ç—å –ø–æ–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –≥—Ä—É–ø–ø –Ω–∞—à–µ–≥–æ –º–∞—Å—Å–∏–≤–∞
                                              # –Ω–µ —Ä–∞–≤–Ω–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –≥—Ä—É–ø–ø –≤ "–∏–¥–µ–∞–ª—å–Ω–æ–º –º–∞—Å—Å–∏–≤–µ"
    i+=1
    if len(dist)<100:                         # –µ—Å–ª–∏ –≤—ã–µ—Ö–∞–ª–∏ –µ—â–µ –Ω–µ –≤—Å–µ 100 —Ç—Ä–∞–º–≤–∞–µ–≤
      dist = np.append(dist,0)+trains[:i]     # –∞–ª–≥–æ—Ä–∏—Ç–º —Ç–æ—Ç –∂–µ, –¥–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤–æ–µ –ø—Ä–æ–π–¥–µ–Ω–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, —Ä–∞–≤–Ω–æ–µ —Å–∫–æ—Ä–æ—Å—Ç–∏ –ø–∞—Ä–æ–≤–æ–∑–∞
    else:
      dist = dist+trains[:i]                  # –µ—Å–ª–∏ —É–∂–µ –≤—ã–µ—Ö–∞–ª–∏ –≤—Å–µ, —Ç–æ –ø—Ä–æ—Å—Ç–∏ –¥–æ–±–∞–≤–ª—è–µ–º –∏—Ö —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —Å–∫–æ—Ä–æ—Å—Ç–∏ –∫ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—é
    for j in range(1,len(dist)):              # –∑–¥–µ—Å—å –∑–∞–º–µ–Ω—è–µ–º —Å–∫–æ—Ä–æ—Å—Ç–∏ –ø–æ–µ–∑–¥–æ–≤ —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º –±–æ–ª—å—à–µ, —á–µ–º —É  –≤–ø–µ—Ä–µ–¥–∏—Å—Ç–æ—è—â–µ–≥–æ –ø–æ–µ–∑–¥–∞ —Å –º–µ–Ω—å—à–µ–π —Å–∫–æ—Ä–æ—Å—Ç—å—é
      if dist[j] > dist[j - 1]:
        dist[j] = dist[j - 1]
        trains[:i][j] = trains[:i][j - 1]
  return i                                    # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç
```
